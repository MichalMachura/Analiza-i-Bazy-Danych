```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import statsmodels.formula.api as sm
import statsmodels.api as sma
import numpy as np
```

```python
data = pd.read_csv('../Orginal Data/beauty/beauty.csv')
data.describe()
#print(data.columns)
```

```python
# assign to another variables
beauty = data.btystdave
evaluation = data.courseevaluation

all_vars = ['tenured', 'profnumber', 'minority', 'age', 'beautyf2upper',
       'beautyflowerdiv', 'beautyfupperdiv', 'beautym2upper',
       'beautymlowerdiv', 'beautymupperdiv', 'btystdave', 'btystdf2u',
       'btystdfl', 'btystdfu', 'btystdm2u', 'btystdml', 'btystdmu', 'class1',
       'class2', 'class3', 'class4', 'class5', 'class6', 'class7', 'class8',
       'class9', 'class10', 'class11', 'class12', 'class13', 'class14',
       'class15', 'class16', 'class17', 'class18', 'class19', 'class20',
       'class21', 'class22', 'class23', 'class24', 'class25', 'class26',
       'class27', 'class28', 'class29', 'class30',
       'didevaluation', 'female', 'formal', 'fulldept', 'lower',
       'multipleclass', 'nonenglish', 'onecredit', 'percentevaluating',
       'profevaluation', 'students', 'tenuretrack', 'blkandwhite',
       'btystdvariance', 'btystdavepos', 'btystdaveneg']
```

```python
# create model
model = sm.ols(formula="courseevaluation ~ btystdave", data=data).fit()
print(model.params)
```

```python
# model description
print(model.summary())
```

```python
# predict values
predicted = model.predict(beauty)
```

```python
# display model
fig1 = go.Figure()

fig1.add_trace(go.Scatter(x=beauty, y=evaluation, mode='markers', name='real'))
fig1.add_trace(go.Scatter(x=beauty, y=predicted, mode='lines', name='predicted'))


fig1.update_layout(
    title="Model ",
    xaxis_title="beauty",
    yaxis_title="evaluation")

fig1.show()
```

Niska wartoÅ›Ä‡ wspÃ³Å‚czynnika przy zmiennej beauty, Å›wiadczy o niewielkiej zaleÅ¼noÅ›ci pomiÄ™dzy tÄ… ziennÄ…, a wartÅ›ciÄ… wyjÅ›ciowÄ… 'evaluation'.
WspÃ³Å‚czynnik 'Intercept' jest wartosciÄ… przesuniecia prostej. Åšwiadczy o przesunieciu danych, czyli niejako o ich nie unormowaniu.

```python
# residuals
r = evaluation - predicted

fig2 = go.Figure()
fig2.add_trace(go.Scatter(y=r))
fig2.update_layout(
    title="Residual",
    xaxis_title="sample number",
    yaxis_title="value of residual")
fig2.show()

print(r.describe())
```

Model jest Åºle dopasowany. Åšwiadczy o tym niska wartoÅ›Ä‡ wspÃ³Å‚czynnika R^2. ZauwaÅ¼alana jest znaczna wariancja danych.

```python
# used variable references
var2 = data[['btystdave','profevaluation']]
```

```python
# create next model
model2 = sm.ols(formula="courseevaluation ~ btystdave + profevaluation", data=data).fit()
print(model2.params)
```

WspÃ³Å‚czynnik 'profevaluation' ma znacznie wiÄ™kszÄ… wartoÅ›Ä‡ niÅ¼ pozostaÅ‚e wspÃ³Å‚czynniki. Åšwiadczy to o duÅ¼ej zaleÅ¼noÅ›ci pomiÄ™dzy tÄ… zmiennÄ…, a wynikiem 'courseevaluation'.
WspÅ‚czynnik 'btystdave' odpowiada za wpÅ‚yw zmiennej beauty na wynik 'courseevaluation'. Jego niska wartoÅ›Ä‡, oznacza niewielkie znaczenie tej zmiennej dla zmiennej wynikowej.
'Intercept' jest wartoÅ›ciÄ… staÅ‚Ä…, opdowiedzialnÄ… za przeseuniÄ™cie prostej regresji.

```python
# model 2 description
print(model2.summary())
```

```python
# predict values for second model
predicted2 = model2.predict(var2)
```

```python
# display data set, model and prediction
fig3 = go.Figure()

fig3.add_trace(go.Scatter3d(x=data.btystdave, y=data.profevaluation, z=data.courseevaluation, mode='markers', name='real'))
fig3.add_trace(go.Scatter3d(x=data.btystdave, y=data.profevaluation, z=predicted2, mode='markers', name='predicted'))

fig3.show()
```

```python
# residual for second model
r2 = evaluation - predicted2

fig4 = go.Figure()
fig4.add_trace(go.Scatter(y=r2))
fig4.update_layout(
    title="Residual",
    xaxis_title="sample number",
    yaxis_title="value of residual")
fig4.show()

print(r2.describe())
```
